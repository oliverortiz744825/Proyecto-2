# Cálculo de la Integral Doble utilizando Montecarlo

Supongamos que deseas calcular la integral doble de una función \(f(x, y)\) en un dominio \(D\) en el plano \(xy\):

$
I=\int\int_D f(x,y) dx dy
$


```python
import numpy as np
import random
import pandas as pd

# Definir la función f(x, y)
def f(x, y):
    return x**2 + y**2

# Definir los límites del dominio
x_min, x_max = 0, 1
y_min, y_max = 0, 1
n_samples = 10000  # Número de puntos aleatorios
x_samples = np.random.uniform(x_min, x_max, n_samples)
y_samples = np.random.uniform(y_min, y_max, n_samples)
integral_estimate = 0

for i in range(n_samples):
    integral_estimate += f(x_samples[i], y_samples[i])

integral_estimate *= (x_max - x_min) * (y_max - y_min) / n_samples

# Crear una tabla con los resultados
data = {
    'Número de muestras': [n_samples],
    'Valor estimado de la integral': [integral_estimate]
}

df = pd.DataFrame(data)
df
```

### Problema 2  
**Problema de simulación de montecarlo en el caso de apostador con el lanzamiento de moneda y poder llegar a ver si es cierto lo de que la moneda es un 50% ganar y un 50% perder.**


>Aqui se va a demostrar que pasa si se tira la moneda apostando cierta cantidad dinero que y porcentaje hay de perdida o ganancia despues de cada tiro, se mmostrara una grafica presentando mostrando como estuvo cambiando el capital despues de cada lanzamiento. 


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Parámetros del juego
capital_inicial = 1000  # Capital inicial del apostador
n_lanzamientos = 100  # Número de lanzamientos de la moneda
apuesta_inicial = 10  # Cantidad apostada en cada lanzamiento

# Función para simular un lanzamiento de moneda (cara = 1, cruz = 0)
def lanzamiento_moneda():
    return np.random.randint(2)

# Función para simular el juego
def simulacion(capital_inicial, n_lanzamientos, apuesta_inicial):
    capital = capital_inicial
    resultados = []
    
    for _ in range(n_lanzamientos):
        resultado_lanzamiento = lanzamiento_moneda()
        
        # Si es cara, el apostador gana y recupera su apuesta
        if resultado_lanzamiento == 1:
            capital += apuesta_inicial
        # Si es cruz, el apostador pierde su apuesta
        else:
            capital -= apuesta_inicial
        
        resultados.append(capital)
    
    return resultados

# Realizar la simulación
resultados = simulacion(capital_inicial, n_lanzamientos, apuesta_inicial)

# Crear un DataFrame para los resultados
df = pd.DataFrame(resultados, columns=["Capital"])

# Graficar el capital a lo largo del tiempo
plt.figure(figsize=(12, 6))
plt.plot(df["Capital"])
plt.title("Evolución del Capital del Apostador")
plt.xlabel("Número de Lanzamientos")
plt.ylabel("Capital")
plt.grid(True)
plt.show()

# Calcular estadísticas
capital_final = df["Capital"].iloc[-1]
probabilidad_ganar = (capital_final - capital_inicial) / (n_lanzamientos * apuesta_inicial)

print(f"Capital Final: ${capital_final}")
print(f"Probabilidad de Ganar por Lanzamiento: {probabilidad_ganar:.2%}")
```
